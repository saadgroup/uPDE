# Burgers Equation

The viscous Burgers equation is a fundamental nonlinear PDE that combines advection and diffusion:

$$\frac{\partial u}{\partial t} + \frac{\partial}{\partial x}\!\left(\frac{u^2}{2}\right) = \nu\,\frac{\partial^2 u}{\partial x^2}$$

It is written in conservation form with flux $F(u) = u^2/2$. The nonlinear advection steepens the wave toward a shock; viscosity $\nu$ regularises it. In the inviscid limit ($\nu=0$) a true discontinuity forms at finite time.

We compare viscous and inviscid cases starting from $u_0 = \sin(x)$.


```python
import numpy as np
import matplotlib.pyplot as plt
import sys; sys.path.insert(0, '..')   # adjust path to upde.py and equations.py
from upde import PDE, PDESystem, Burgers
```

```python
x = np.linspace(0, 2*np.pi, 512)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

for ax, nu, label in [
    (axes[0], 0.05,  'Viscous  ν=0.05'),
    (axes[1], 0.001, 'Nearly inviscid  ν=0.001'),
]:
    eq = Burgers('u', x=x, viscosity=nu)
    eq.set_bc(kind='periodic')
    eq.set_ic(np.sin(x))

    sol = PDESystem([eq]).solve(t_span=(0, 3), method='RK45',
                                t_eval=np.linspace(0, 3, 5),
                                rtol=1e-4, atol=1e-6)
    for k, t in enumerate(sol.t):
        ax.plot(x, sol.u[:, k], label=f't={t:.1f}')
    ax.set_xlabel('x'); ax.set_ylabel('u')
    ax.set_title(label)
    ax.legend(fontsize=8)

plt.tight_layout(); plt.show()
```
